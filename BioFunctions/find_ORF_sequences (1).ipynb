```python
import pprint
from Bio import SeqIO
input_file = "dna2.fasta"
sequences = [i for i in SeqIO.parse(input_file, "fasta")]
```

```python
def find_ORF_sequences(sequences):
    #Initialize empty dicts for populating later
    ORFs_reading1={}
    ORFs_reading2={}
    ORFs_reading3={}
    
    #Reading frame 1
    for i in range(len(sequences)):
        current_sequence = str(sequences[i].seq)
        current_sequence_id = sequences[i].id
        start_j = -1
        stop_j = -1
        start_codon_found = 0
        for j in range(len(current_sequence)-3):
            #Partition sequence in three codons
            j*=3
            #Find start codon ATG
            if start_codon_found == 0:
                if current_sequence[j:j+3] == "ATG":
                    start_j = j
                    #Stop searching for start codon
                    start_codon_found = 1
            #If start codon is found, search for stop codon
            if start_codon_found == 1:
                if (current_sequence[j:j+3] == "TAA") or (current_sequence[j:j+3] == "TAG") or (current_sequence[j:j+3] == "TGA"):
                    #Restart search for new ORF
                    start_codon_found = 0
                    stop_j = j+3
                    ORF_full = current_sequence[start_j:stop_j]
                    ORF_length = len(ORF_full)
                    ORF_start_stop = [start_j, stop_j]
                    ORF = [ORF_full, ORF_length, ORF_start_stop]
                    if sequences[i].name in ORFs_reading1.keys():
                        ORFs_reading1[sequences[i].name].append(ORF)  
                    else:
                        ORFs_reading1[sequences[i].name] = [ORF]
    #pprint.pprint(ORFs_reading1)
    
    #Reading frame 2
    for i in range(len(sequences)):
        current_sequence = str(sequences[i].seq)
        current_sequence = current_sequence[1:]
        current_sequence_id = sequences[i].id
        start_j = -1
        stop_j = -1
        start_codon_found = 0
        for j in range(len(current_sequence)-3):
            #Partition sequence in three codons
            j*=3
            #Find start codon ATG
            if start_codon_found == 0:
                if current_sequence[j:j+3] == "ATG":
                    start_j = j
                    #Stop searching for start codon
                    start_codon_found = 1
            #If start codon is found, search for stop codon
            if start_codon_found == 1:
                if (current_sequence[j:j+3] == "TAA") or (current_sequence[j:j+3] == "TAG") or (current_sequence[j:j+3] == "TGA"):
                    #Restart search for new ORF
                    start_codon_found = 0
                    stop_j = j+3
                    ORF_full = current_sequence[start_j:stop_j]
                    ORF_length = len(ORF_full)
                    ORF_start_stop = [start_j, stop_j]
                    ORF = [ORF_full, ORF_length, ORF_start_stop]
                    if sequences[i].name in ORFs_reading2.keys():
                        ORFs_reading2[sequences[i].name].append(ORF)  
                    else:
                        ORFs_reading2[sequences[i].name] = [ORF]
    #pprint.pprint(ORFs_reading2)
    
    #Reading frame 3
    for i in range(len(sequences)):
        current_sequence = str(sequences[i].seq)
        current_sequence = current_sequence[2:]
        current_sequence_id = sequences[i].id
        start_j = -1
        stop_j = -1
        start_codon_found = 0
        for j in range(len(current_sequence)-3):
            #Partition sequence in three codons
            j*=3
            #Find start codon ATG
            if start_codon_found == 0:
                if current_sequence[j:j+3] == "ATG":
                    start_j = j
                    #Stop searching for start codon
                    start_codon_found = 1
            #If start codon is found, search for stop codon
            if start_codon_found == 1:
                if (current_sequence[j:j+3] == "TAA") or (current_sequence[j:j+3] == "TAG") or (current_sequence[j:j+3] == "TGA"):
                    #Restart search for new ORF
                    start_codon_found = 0
                    stop_j = j+3
                    ORF_full = current_sequence[start_j:stop_j]
                    ORF_length = len(ORF_full)
                    ORF_start_stop = [start_j, stop_j]
                    ORF = [ORF_full, ORF_length, ORF_start_stop]
                    if sequences[i].name in ORFs_reading3.keys():
                        ORFs_reading3[sequences[i].name].append(ORF)  
                    else:
                        ORFs_reading3[sequences[i].name] = [ORF]
    #pprint.pprint(ORFs_reading3)
    
    return ORFs_reading1, ORFs_reading2, ORFs_reading3
```

```python
#Example of returned dictionaries
ORF1, ORF2, ORF3 = find_ORF_sequences(sequences)
```
